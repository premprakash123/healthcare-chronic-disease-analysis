```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/ckd_clean_initial.csv")

df.info()

## Check for missing values
df.isnull().sum()
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
    print("----------------")
```

```python
##Replace all invalid values
df.replace("?", np.nan, inplace=True)
df.replace("\\t?", np.nan, inplace=True)
df.replace(" ", np.nan, inplace=True)
```

```python
##Check for missing values again
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
##Convert numeric columns
numeric_cols = [
    'age',
    'bp',
    'sg',
    'al',
    'su',
    'bgr',
    'bu',
    'sc',
    'sod',
    'pot',
    'hemo',
    'pcv',
    'wc',
    'rc'
]
```

```python
print(df.columns.tolist())
```

```python
## remove extra quotes from column names
df.columns = df.columns.str.replace("'", "", regex=False)
print(df.columns.tolist())
```

```python
numeric_cols = [
    'age',
    'bp',
    'sg',
    'al',
    'su',
    'bgr',
    'bu',
    'sc',
    'sod',
    'pot',
    'hemo',
    'pcv',
    'wbcc',
    'rbc'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
numeric_cols = [
    'age',
    'bp',
    'sg',
    'al',
    'su',
    'bgr',
    'bu',
    'sc',
    'sod',
    'pot',
    'hemo',
    'pcv',
    'wbcc',
    'rbcc'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = ['pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane','class']

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
##remove rbc column as it is redundant with pcv
df.drop(columns=['rbc'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/ckd_cleaned.csv", index=False)
```
