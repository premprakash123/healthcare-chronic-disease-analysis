```python
# ==========================================
# Import Required Libraries
# ==========================================

import pandas as pd
import numpy as np

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model_results = {

    "Model": [
        
        "Logistic Regression",
        
        "Decision Tree",
        
        "Random Forest",
        
        "KNN",
        
        "Optimized Random Forest"
    ],

    "Accuracy": [
        
        1.00,
        
        1.00,
        
        0.99,
        
        0.97,
        
        0.99
    ],

    "Precision": [
        
        1.00,
        
        1.00,
        
        0.99,
        
        0.98,
        
        0.99
    ],

    "Recall": [
        
        1.00,
        
        1.00,
        
        0.98,
        
        0.97,
        
        0.98
    ],

    "F1_Score": [
        
        1.00,
        
        1.00,
        
        0.99,
        
        0.98,
        
        0.99
    ]
}
```

```python
# ==========================================
# Create Comparison DataFrame
# ==========================================

comparison_df = pd.DataFrame(
    
    model_results
)

comparison_df
```

```python
# ==========================================
# Display Best Performing Model
# ==========================================

best_model = comparison_df.sort_values(
    
    by="Accuracy",
    
    ascending=False
)

print("Best Performing Model:\n")

print(best_model.head(1))
```

```python
# ==========================================
# Accuracy Comparison Plot
# ==========================================

plt.figure(figsize=(10,6))

sns.barplot(
    
    x='Model',
    
    y='Accuracy',
    
    data=comparison_df
)

plt.title("Healthcare Model Accuracy Comparison")

plt.ylim(0.8, 1.0)

plt.xlabel("Machine Learning Models")

plt.ylabel("Accuracy Score")

plt.show()
```

```python
# ==========================================
# Precision Comparison Plot
# ==========================================

plt.figure(figsize=(10,6))

sns.barplot(
    
    x='Model',
    
    y='Precision',
    
    data=comparison_df
)

plt.title("Healthcare Model Precision Comparison")

plt.ylim(0.8, 1.0)

plt.xlabel("Machine Learning Models")

plt.ylabel("Precision Score")

plt.show()
```

```python
# ==========================================
# Recall Comparison Plot
# ==========================================

plt.figure(figsize=(10,6))

sns.barplot(
    
    x='Model',
    
    y='Recall',
    
    data=comparison_df
)

plt.title("Healthcare Model Recall Comparison")

plt.ylim(0.8, 1.0)

plt.xlabel("Machine Learning Models")

plt.ylabel("Recall Score")

plt.show()
```

```python
# ==========================================
# F1-Score Comparison Plot
# ==========================================

plt.figure(figsize=(10,6))

sns.barplot(
    
    x='Model',
    
    y='F1_Score',
    
    data=comparison_df
)

plt.title("Healthcare Model F1-Score Comparison")

plt.ylim(0.8, 1.0)

plt.xlabel("Machine Learning Models")

plt.ylabel("F1-Score")

plt.show()
```
